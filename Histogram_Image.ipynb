```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from google.colab import files

# Upload image (downloaded from Google)
uploaded = files.upload()

# Load image (replace name if different)
image = cv2.imread("pexels-suzyhazelwood-1510392.jpg")

# Convert BGR â†’ RGB for display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert to grayscale
gray = rgb2gray(image_rgb)
gray = (gray * 255).astype(np.uint8)

# ONE single figure
plt.figure(figsize=(18,5))

#Original Image
plt.subplot(1,3,1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")

#Color Histogram
plt.subplot(1,3,2)
plt.title("Color Image Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Count")

plt.plot(cv2.calcHist([image],[0],None,[256],[0,256]), color='blue', label='Blue')
plt.plot(cv2.calcHist([image],[1],None,[256],[0,256]), color='green', label='Green')
plt.plot(cv2.calcHist([image],[2],None,[256],[0,256]), color='red', label='Red')
plt.legend()

#Grayscale Histogram
plt.subplot(1,3,3)
plt.title("Grayscale Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Count")

plt.hist(gray.ravel(), bins=256, range=[0,256], color='gray', label='Gray')
plt.legend()

plt.tight_layout()
plt.show()
```
