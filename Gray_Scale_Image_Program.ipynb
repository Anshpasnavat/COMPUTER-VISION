```python
# Import required libraries
from skimage import io, color
import matplotlib.pyplot as plt

import numpy as np
```

```python
# Import files module to upload image in Google Colab
# Upload image from local system
from google.colab import files
uploaded = files.upload()
```

```python
# Read the uploaded image using skimage
# list(uploaded.keys())[0] gets the uploaded file name
image = io.imread('pexels-suzyhazelwood-1510392.jpg')
```

```python
# Convert the RGB image to grayscale
gray_image = color.rgb2gray(image)
```

```python

# Display the original image
plt.imshow(image)
plt.title("Loaded Image")
plt.axis('off')
plt.show()
```

```python
# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python

# Save the grayscale image to disk
gray_image_uint8 = (gray_image * 255).astype(np.uint8)
io.imsave("grayscale_image.jpg", gray_image_uint8)
```
